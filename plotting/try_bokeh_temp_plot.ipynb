```python
from bokeh.models.widgets import Panel, Tabs
from bokeh.plotting import figure
from bokeh.resources import CDN
from bokeh.embed import file_html
```

```python
from bokeh.io import show, output_notebook
output_notebook()
```

```python
p1 = figure(plot_width=300, plot_height=300)
p1.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=20, color="navy", alpha=0.5)
tab1 = Panel(child=p1, title="temp")

p2 = figure(plot_width=300, plot_height=300)
p2.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=3, color="navy", alpha=0.5)
tab2 = Panel(child=p2, title="hum")

tabs = Tabs(tabs=[ tab1, tab2 ])

```

```python
tempseries = pd.read_csv('download.csv',header=None, names=['timestamp','temperature'], parse_dates=True, index_col=0)
```

```python
tempseries.head()
```

```python
from bokeh.models import ColumnDataSource, HoverTool
```

```python
hover = HoverTool(tooltips=[('temp','@temperature{0.0} C'),
                            ('time','@timestamp{%d-%b %H:%M}')],
                  formatters={'timestamp':'datetime'},
                  mode='vline'
                 )

f = figure(height=300, width=800, 
           x_axis_type="datetime", 
           toolbar_location="above",
           title='logging data',
           tools = [hover])

f.xaxis[0].axis_label = 'Date'
f.yaxis[0].axis_label = 'Temperature (C)'

f.line(x='timestamp',y='temperature',source=ColumnDataSource(tempseries))
show(f)
```

```python
hover1 = HoverTool(tooltips=[('temp','@temperature{0.0} C'),
                            ('time','@timestamp{%d-%b %H:%M}')],
                  formatters={'timestamp':'datetime'},
                  mode='vline'
                 )
f1 = figure(height=300, width=800, 
           x_axis_type="datetime", 
           toolbar_location="above",
           tools = [hover1])
f1.xaxis[0].axis_label = 'Date'
f1.yaxis[0].axis_label = 'Temperature (C)'
f1.line(x='timestamp',y='temperature',source=ColumnDataSource(tempseries))
tab1 = Panel(child=f1, title="temp")

hover2 = HoverTool(tooltips=[('hum','@temperature{0.0} %'),
                            ('time','@timestamp{%d-%b %H:%M}')],
                  formatters={'timestamp':'datetime'},
                  mode='vline'
                 )
f2 = figure(height=300, width=800, 
           x_axis_type="datetime", 
           toolbar_location="above",
           tools = [hover2])
f2.xaxis[0].axis_label = 'Date'
f2.yaxis[0].axis_label = 'Humidity (%)'
f2.line(x='timestamp',y='temperature',source=ColumnDataSource(tempseries), color='orange')
tab2 = Panel(child=f2, title="hum")

tabs = Tabs(tabs=[ tab1, tab2 ])



show(tabs)
```
